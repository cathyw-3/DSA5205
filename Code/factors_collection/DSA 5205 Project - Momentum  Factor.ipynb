```python
data = pd.read_csv('Raw_Stock_Data (1).csv')
```

```python
data=data.rename(columns={'Adj Close':'Adj_Close'})
```

To add the additional columns for momentum factors into the dataframe

```python
data[['relative_strength_1m',"relative_strength_2m", "relative_strength_3m", "relative_strength_6m",'relative_strength_12m']] = None
```

```python
data
```

To get the total numbers of stockets in this data

```python

tickets = data["Index"].unique()
tickets[0]
```

```python
len(tickets)
```

To filter the dataset with date from 2021-01 to 2022-07

```python
data[(data['Index']=='FTNT') & (data['Date'] > '2020-12-31')]
```

To calculate the momentun factors for each stock

```python
for b in range(0,len(tickets)):
    for i in range(253+b*649,(648+b*649)+1):
        data.relative_strength_1m[i]= (data.Adj_Close[i] - data.Adj_Close[i-21])/ data.Adj_Close[i-21]
        data.relative_strength_2m[i]= (data.Adj_Close[i] - data.Adj_Close[i-21*2])/ data.Adj_Close[i-21*2]
        data.relative_strength_3m[i]= (data.Adj_Close[i] - data.Adj_Close[i-21*3])/ data.Adj_Close[i-21*3]
        data.relative_strength_6m[i]= (data.Adj_Close[i] - data.Adj_Close[i-21*6])/ data.Adj_Close[i-21*6]
        data.relative_strength_12m[i]= (data.Adj_Close[i] - data.Adj_Close[i-21*12])/ data.Adj_Close[i-21*12]
            
```

```python
tables=[]
for x in tickets:
    table = data[(data['Index']==x) & (data['Date'] > '2020-12-31')]
    tables.append(table)
```

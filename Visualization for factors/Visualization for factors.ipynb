```python
import pandas as pd
import plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data = pd.read_csv('./factors_returns.csv')
```

```python
data
```

```python
data.columns,len(list(data))
```

## Plot scatterplot

```python
line_list_Momentum=[]
for i in range(1,4):
   i = go.Scatter(x=data.Date, y=data[list(data)[i]],name = list(data)[i])
   line_list_Momentum.append(i)

fig_Momentum = go.Figure(data=line_list_Momentum)
fig_Momentum.update_layout(title='Momentum factors', xaxis_title='Date', yaxis_title='Momentum factors value')
fig_Momentum.show()

line_list_Turnover=[]
for t in range(4,7):
   t = go.Scatter(x=data.Date, y=data[list(data)[t]],name = list(data)[t])
   line_list_Turnover.append(t)

fig_Turnover = go.Figure(data=line_list_Turnover)
fig_Turnover.update_layout(title='Turnover factors', xaxis_title='Date', yaxis_title='Turnover factors value')
fig_Turnover.show()

line_list_Volatility=[]
for v in range(7,13):
   v = go.Scatter(x=data.Date, y=data[list(data)[v]],name = list(data)[v])
   line_list_Volatility.append(v)

fig_Volatility = go.Figure(data=line_list_Volatility)
fig_Volatility.update_layout(title='Volatility factors', xaxis_title='Date', yaxis_title='Volatility factors value')
fig_Volatility.show()

line_list_Financial_Quality_Factor=[]
for s in range(13,15):
   s = go.Scatter(x=data.Date, y=data[list(data)[s]],name = list(data)[s])
   line_list_Financial_Quality_Factor.append(s)

fig_Financial_Quality_Factor = go.Figure(data=line_list_Financial_Quality_Factor)
fig_Financial_Quality_Factor.update_layout(title='Financial Quality factor', xaxis_title='Date', yaxis_title='Finanical Quality factors value')
fig_Financial_Quality_Factor.show()

line_list_Value_Factor=[]
for g in range(15,17):
   g = go.Scatter(x=data.Date, y=data[list(data)[g]],name = list(data)[g])
   line_list_Value_Factor.append(g)

fig_Value_Factor = go.Figure(data=line_list_Value_Factor)
fig_Value_Factor.update_layout(title='Value factor', xaxis_title='Date', yaxis_title='Value factor value')
fig_Value_Factor.show()
```

## Plot Histogram plot

```python
fig = make_subplots(rows=1, cols=3, shared_yaxes=False, subplot_titles=(list(data)[1], list(data)[2], list(data)[3]))

for a in range(1,4):
   a = fig.add_trace(go.Histogram(x=data[list(data)[a]], name = list(data)[a],xbins = dict(size = 0.001)),row=1, col=a)

fig.update_layout(title='Momentum factors',yaxis_title='Frequency',bargap=0.2, bargroupgap=0.1,height=600,width=1400).show()
```

```python
fig1 = make_subplots(rows=1, cols=3, shared_yaxes=False, subplot_titles=(list(data)[4], list(data)[5], list(data)[6]))

for t in range(4,7):
   t = fig1.add_trace(go.Histogram(x=data[list(data)[t]], name = list(data)[t],xbins = dict(size = 0.001)),row=1, col=t-3)

fig1.update_layout(title='Turnover factors',yaxis_title='Frequency',bargap=0.2, bargroupgap=0.1,height=600,width=1400).show()
```

```python
fig3 = make_subplots(rows=2, cols=3, shared_yaxes=False, y_title = 'Frequency',subplot_titles=(list(data)[7], list(data)[8], list(data)[9],list(data)[10], list(data)[11], list(data)[12]))

for v in range(7,10):
   v = fig3.add_trace(go.Histogram(x=data[list(data)[v]], name = list(data)[v],xbins = dict(size = 0.001)),row=1, col=v-6)
for b in range(10,13):
   b = fig3.add_trace(go.Histogram(x=data[list(data)[b]], name = list(data)[b],xbins = dict(size = 0.01)),row=2, col=b-9)

fig3.update_layout(title='Volatility factors',bargap=0.2, bargroupgap=0.1,height=600,width=1400).show()
```

```python
fig4 = make_subplots(rows=1, cols=2, shared_yaxes=False, subplot_titles=(list(data)[13], list(data)[14]))

for f in range(13,15):
   f = fig4.add_trace(go.Histogram(x=data[list(data)[f]], name = list(data)[f],xbins = dict(size = 0.001)),row=1, col=f-12)

fig4.update_layout(title='Financial Quality Factor',yaxis_title='Frequency',bargap=0.2, bargroupgap=0.1,height=600,width=1400).show()
```
